```python
def loan_bucket (loan_amnt):
    if loan_amnt<5000:
        return '5kbucket'
    elif loan_amnt<=10000:
        return '5k-10kbucket'
    elif loan_amnt<=15000:
        return '10-15kbucket'
    elif loan_amnt<=20000:
        return '10k-20kbucket'
    elif loan_amnt<=25000:
        return '20k-25kbucket'
    elif loan_amnt<=30000:
        return '25k-30kbucket'
    else:
        return '30k-35kbucket'
    
     
       
  
```

```python
df_only_defaulter_data=df[(df['non_default_status']=='False')]
df_only_defaulter_data['Loan_amount_bucket']=df_only_defaulter_data['loan_amnt'].apply(loan_bucket)
df_only_defaulter_data
```

##### from the below viz. it is understood people who opt for loan amount 5k to 10k are more likely to default the loan refund 

```python
df_only_defaulter_group_loan_bucket=df_only_defaulter_data.groupby('Loan_amount_bucket')['id'].count()
loan_bucket_count=pd.DataFrame(df_only_defaulter_group_loan_bucket).reset_index()
plt.bar(loan_bucket_count['Loan_amount_bucket'],loan_bucket_count['id'])
plt.xlabel('Loan Bucket')
plt.ylabel('def_count')
plt.title('Loan amount wise  Defaulter')
plt.xticks(rotation=45)
plt.show()



```
